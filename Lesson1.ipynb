## Исследование набора данных сортов изюма Кечимен и Бесни

### Подключение библиотеки NumPy и загрузка данных

```python
import numpy as np
data_path = "datasets/Raisin/Raisin_Dataset.data"
data = np.genfromtxt(data_path, delimiter=";")
print(data)
```

### Тип переменной и форма (shape)

```python
print ( "Data type : ", type(data) )
print ( "Data shape : ", data.shape )
print ( data[:10] )
```

### Получение типа набора данных, строки, элемента

```python
data1 = np.genfromtxt(data_path, delimiter=";", dtype=None)
print('Shape of the dataset:', data1.shape)
print('Dataset type:', type(data1))
print('A single row of the dataset is type of:', type(data1[0]))
print('Types of elements:', type(data1[0][1]), type(data1[0][4]))
print('Dataset:')
print(data1)
```

### Указание типа столбцов при загрузке данных

```python
dt = np.dtype("i, f8, f8, f8, i, f8, f8, U30")
data2 = np.genfromtxt(data_path, delimiter=";", dtype=dt)
print('Shape of the dataset:', data2.shape)
print('Dataset type:', type(data2))
print('A single row of the dataset is type of:', type(data2[0]))
print('Types of elements:', type(data2[0][1]), type(data2[0][4]))
print('Dataset slice:')
print(data2[:10])
```

### Построение графиков с использованием Matplotlib

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

Perimeter = []
MajorAxisLength = []
MinorAxisLength = []
Eccentricity = []
Extent = []
Area = []
ConvexArea = []

for dot in data:
    Area.append(dot[0])
    Perimeter.append(dot[1])
    MajorAxisLength.append(dot[2])
    MinorAxisLength.append(dot[3])
    Eccentricity.append(dot[4])
    ConvexArea.append(dot[5])
    Extent.append(dot[6])

print("Min Area: ", min(Area))
print("Min Perimeter: ", min(Perimeter))
print("Min MajorAxisLength: ", min(MajorAxisLength))
print("Min MinorAxisLength: ",min( MinorAxisLength))
print("Min Eccentricity: ", min(Eccentricity))
print("Min ConvexArea: ", min(ConvexArea))
print("Min Extent: ", min(Extent))

print("Max Area: ", max(Area))
print("Max Perimeter: ", max(Perimeter))
print("Max MajorAxisLength: ", max(MajorAxisLength))
print("Max MinorAxisLength: ",max( MinorAxisLength))
print("Max Eccentricity: ", max(Eccentricity))
print("Max ConvexArea: ", max(ConvexArea))
print("Max Extent: ", max(Extent))

plt.figure(1)
Kecimen, = plt.plot(MajorAxisLength[0:450], MinorAxisLength[0:450], 'g^', label='Kecimen')
Besni, = plt.plot(MajorAxisLength[451:900], MinorAxisLength[451:900], 'bs', label='Besni')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('MajorAxisLength')
plt.ylabel('MinorAxisLength')

plt.show()

plt.figure(2)
Kecimen, = plt.plot(Perimeter[0:450], MinorAxisLength[0:450], 'g^', label='Kecimen')
Besni, = plt.plot(Perimeter[451:900], MinorAxisLength[451:900], 'bs', label='Besni')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Perimeter')
plt.ylabel('MinorAxisLength')

plt.show()

plt.figure(3)
Kecimen, = plt.plot(Perimeter[0:450], Eccentricity[0:450], 'g^', label='Kecimen')
Besni, = plt.plot(Perimeter[451:900], Eccentricity[451:900], 'bs', label='Besni')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Perimeter')
plt.ylabel('Eccentricity')

plt.show()

plt.figure(4)
Kecimen, = plt.plot(Perimeter[0:450], Extent[0:450], 'g^', label='Kecimen')
Besni, = plt.plot(Perimeter[451:900], Extent[451:900], 'bs', label='Besni')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Perimeter')
plt.ylabel('Extent')

plt.show()

plt.figure(5)
Kecimen, = plt.plot(Eccentricity[0:450], Extent[0:450], 'g^', label='Kecimen')
Besni, = plt.plot(Eccentricity[451:900], Extent[451:900], 'bs', label='Besni')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Eccentricity')
plt.ylabel('Extent')

plt.show()

plt.figure(6)
Kecimen, = plt.plot(ConvexArea[0:450], Area[0:450], 'g^', label='Kecimen')
Besni, = plt.plot(ConvexArea[451:900], Area[451:900], 'bs', label='Besni')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('ConvexArea')
plt.ylabel('Area')

plt.show()

plt.figure(7)
Kecimen, = plt.plot(Perimeter[0:450], MajorAxisLength[0:450], 'g^', label='Kecimen')
Besni, = plt.plot(Perimeter[451:900], MajorAxisLength[451:900], 'bs', label='Besni')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Perimeter')
plt.ylabel('MajorAxisLength')

plt.show()
```
